```python
import sys
sys.path.insert(0,'../')
import pymce as mce

import time
import numpy as np

from sklearn import neighbors 
from sklearn.ensemble import IsolationForest
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA,TruncatedSVD,NMF,FastICA,KernelPCA,IncrementalPCA
from sklearn.metrics import roc_auc_score

import matplotlib.pylab as plt
from matplotlib import gridspec
%matplotlib inline
```

```python
agg = AgglomerativeClustering(n_clusters=2,
            affinity='euclidean', connectivity=None,
            compute_full_tree='auto', linkage='ward')

def clustering(z_mu):
    num = z_mu.shape[0]
    n_divide = int(num/4999)+1
    y = np.zeros(num)
    for inds in np.array_split(np.arange(num), n_divide):
        y[inds] = agg.fit_predict(z_mu[inds,:])
    return y

metrics = mce.metrics
```

```python
def synt_mix(i_sig, n_ftrs,x=None,n_sig=11,n_inlier=1000,n_outlier=50,sigma = 0.2,n1 = 0.02,n2 = 0.01,n3 = 0.02,n4 = 0.01):
    main_data = {i_sig:n_inlier}

    if x is None:
        x = np.linspace(0,1,n_ftrs)
    X = []
    y = []
    for key,value in main_data.iteritems():
        for _ in range(value):
            Xp = mce.signal(key,x,sigma,n1,n2,n3,n4)
            X.append(Xp)
            y.append(key)

#     for i in range(1,n_sig):
    for i in range(10,11):
        if i!=i_sig:
            for j in range(n_outlier):
                Xp = mce.signal(i,x,sigma,n1,n2,n3,n4)
                X.append(Xp)
                y.append(i)
            
    return np.array(X),np.array(y)
```

```python
i_sig = 2
n_ftrs = 1000
noise = 0.1
scl = 0.0
sft = 0.0

x = np.linspace(0,1,n_ftrs)

X, y = synt_mix(i_sig,n_ftrs,x=x,sigma = noise,n1 = scl,n2 = sft,n3 = scl,n4 = sft)

gs = gridspec.GridSpec(1, 2)
plt.figure(figsize=(8,3)) 
ax1 = plt.subplot(gs[0, 0])
ax2 = plt.subplot(gs[0, 1])
ax1.set_title('Inliers')
ax2.set_title('Outliers')

inliers = X[y==i_sig]
outliers = X[y!=i_sig]
outliers_y = y[y!=i_sig]
for i in range(0,1,3):
    ax1.plot(inliers[i],'b')
    ax2.plot(outliers[i],mce.COLORS[outliers_y[i]])
    
# plt.subplots_adjust(hspace=0.3,left=0.1, right=0.9, top=0.9, bottom=0.1)
# plt.savefig('x_max=1_d=1000.jpg')
```

```python
nn_en = [n_ftrs, n_ftrs//2, 2]
nn_de = [2, n_ftrs//2, n_ftrs]
network_architecture = [nn_en,nn_de]

dim_rs ={'AE':'AE','VAE':'VAE', 
           'PCA':PCA(n_components=2),
           'NMF':NMF(n_components=2), 
           'FastICA':FastICA(n_components=2, max_iter=1000)}

# dim_rs ={'AE':'AE'}

for dim_r, value in dim_rs.iteritems():
    
    print '------------------- '+dim_r+' --------------------'

    if dim_r=='VAE' or dim_r=='AE':
        splitter = mce.Splitter(X, value, network_architecture, clustering)
    else:
        splitter = mce.Custom_Splitter(X, mce.sk_convert(value), clustering)

    b_time = time.time()

    for i in range(1):
        if dim_r=='VAE' or dim_r=='AE':
            splitter.split(1,verbose=0,training_epochs=20)
        else:
            splitter.split(1,verbose=0)
        outliers = mce.outliers(X,splitter,metrics)
        
    tenth_outliers = mce.norm_ensemble(outliers,0.1)
    max_outliers = mce.max_ensemble(outliers)
    
    s2 = roc_auc_score(y!=i_sig, tenth_outliers)
    s3 = roc_auc_score(y!=i_sig, max_outliers) 

    lof = neighbors.LocalOutlierFactor(n_neighbors=10)
    lof.fit(X)    
    s1 = roc_auc_score(y!=i_sig, -lof.negative_outlier_factor_)

    print 'LOF:{:4.2f}, 01: {:4.2f}, max: {:4.2f}'.format(s1,s2,s3)

#     for metr in metrics:
#         print metr,roc_auc_score(y!=i_sig, outliers[metr])
        
    print '-------------------------------------------------'
```
