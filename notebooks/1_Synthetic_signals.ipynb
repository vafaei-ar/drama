## Synthetic data 
There are some pre-defined signals and outlier simulator to evaluated **DRAMA**.

```python
import drama as drm
import numpy as np

import matplotlib.pylab as plt
from matplotlib import gridspec
%matplotlib inline
```

## Initiation

```python
# Number of features
n_ftrs = 100

# Domain of signal generator function
x = np.linspace(0,1,n_ftrs)
```

## Signals

```python
sigma = 0.0
n1,n2,n3,n4 = 0,0,0,0

gs = gridspec.GridSpec(2, 5)
plt.figure(figsize=(10, 4)) 

shapes = {}
for i in range(10):
    ax = plt.subplot(gs[i%2, i//2])
    ax.set_xticks([])
    ax.set_yticks([])
    y = drm.signal(i+1,x,sigma,n1,n2,n3,n4)
    ax.plot(y,'b',lw=2)
    ax.set_ylim(-0.1,1.1)
    ax.set_title(str(i+1))

plt.subplots_adjust(bottom=0.05, top=0.9, left=0.05, right=0.95)
# plt.savefig('shapes.jpg',dpi=300)
```

## signal perturbations

```python
clrs = ['r','g','b']
sigma = 0.2
n1,n2,n3,n4 = 0.05,0.05,0.05,0.05

gs = gridspec.GridSpec(2, 5)
plt.figure(figsize=(10, 4)) 

shapes = {}
for i in range(10):
    ax = plt.subplot(gs[i%2, i//2])
    ax.set_xticks([])
    ax.set_yticks([])
    for j in range(3):
        y = drm.signal(i+1,x,sigma,n1,n2,n3,n4)
        ax.plot(y,clrs[j],lw=1)
    ax.set_title(str(i+1))

plt.subplots_adjust(bottom=0.05, top=0.9, left=0.05, right=0.95)
# plt.savefig('shapes_purt.jpg',dpi=300)
```

## event modulation

```python
sigma = 0.0
n1,n2,n3,n4 = 0,0,0,0

gs = gridspec.GridSpec(2, 5)
plt.figure(figsize=(10, 4)) 

shapes = {}
for i in range(10):
    ax = plt.subplot(gs[i%2, i//2])
    ax.set_xticks([])
    ax.set_yticks([])
    y = drm.signal(i+1,x,sigma,n1,n2,n3,n4)
    ax.plot(y,'b',lw=2)
    y = drm.event_sig(y)
    ax.plot(y,'r',lw=2)
    ax.set_title(str(i+1))

plt.subplots_adjust(bottom=0.05, top=0.9, left=0.05, right=0.95)
# plt.savefig('shapes_event.jpg',dpi=300)
```
